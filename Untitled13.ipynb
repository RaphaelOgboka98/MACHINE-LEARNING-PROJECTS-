```python
import os
import pandas as pd
import json
```

```python
data_path='/content/gdrive/MyDrive/myData'
```

```python
files=os.listdir(data_path)
```

```python
print(files)
```

```python
Dataframes={}

for file in files:
  file_path=os.path.join(data_path,file)


  if file.endswith('.csv'):
    df=pd.read_csv(file_path)
    Dataframes[file]=df
  elif file.endswith('.json'):
    with open(file_path,'r') as f:
      data=json.load(f)
      df=pd.DataFrame(data)
      Dataframes[file]=df
```
